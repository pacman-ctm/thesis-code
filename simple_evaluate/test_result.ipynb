```python
def get_caption(image_id):
    for item in test_data['annotations']:
        if item['image_id'] == image_id:
            print(item['caption'])
get_caption(1)
```

```python
def image_captioning(image_id: int):
    print('image_id = ', image_id)
    image_name = 'public-test-images/' + imgid2imgname[image_id]
    image = Image.open(image_name).convert('RGB')
    plt.imshow(image)

    # Print caption
    print('test')
    get_caption(image_id)
    print('res: ', [item['caption'] for item in res_data if item['image_id'] == image_id])
    print('vit: ', [item['caption'] for item in vit_data if item['image_id'] == image_id])
    print('grit: ', grit_data[str(image_id)])
```
